```python
trace = go.Scatter(
    x = [1,2,3,4,5],
    y = [10,20,30,20,10],
    mode = 'markers + lines',
    text=["A","B","C","D","E"],
    marker= dict(size= 14,
                line= dict(width=1),
                color= "red",
                opacity= 0.3
                ),
)

data = [trace]
#plotly.offline.iplot(data)
go.Figure(data).show()
```
